### Onehot encoding

Onehot encoding, not dependent on label encoding. It can directly convert the categorical data into numerical

```python
data.head()
```

```python
from sklearn import preprocessing # library
One_hot = preprocessing.OneHotEncoder() # create instance
One_hot.fit(data[['Gender']]) # specify the column for dummy variables
```

```python
onehotlabels = One_hot.transform(data[['Gender']]).toarray()
```

```python
b = pd.DataFrame(onehotlabels , columns = ['female','male'])
b.head()
```

```python
data = pd.concat([data,b],axis=1)
data.head(4)
```

```python
data = data.drop(['Gender','female'], axis = 1) 

data.head()
```
