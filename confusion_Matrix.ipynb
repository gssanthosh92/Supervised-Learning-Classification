# Confusion Matrix

A confusion matrix is a summary of prediction results on a classification problem.

### Binomial Classification

### Definition of the Terms:
• Positive (P) : Observation is positive 

• Negative (N) : Observation is not positive 

• True Positive (TP) : Observation is positive, and is predicted to be positive.

• False Negative (FN) : Observation is positive, but is predicted negative.

• True Negative (TN) : Observation is negative, and is predicted to be negative.

• False Positive (FP) : Observation is negative, but is predicted positive.

## Accuracy :    

Accuracy is measure of overall performance of the model. However, It assumes equal costs for both kinds of errors.

Accuracy = (TP + TN)/(TP + TN + FP + FN)

## Recall :

Its is also called as sensitivity . Recall can be defined as the ratio of the total number of correctly classified positive examples divide to the total number of positive examples. High Recall indicates the class is correctly recognized (small number of FN).

Recall = (TP)/(TP+FN)

## Precision :

To get the value of precision we divide the total number of correctly classified positive examples by the total number of predicted positive examples. High Precision indicates an example labeled as positive is indeed positive (small number of FP).

Precision = TP / (TP+FP)

High recall, low precision:This means that most of the positive examples are correctly recognized (low FN) but there are a lot of false positives.

Low recall, high precision:This shows that we miss a lot of positive examples (high FN) but those we predict as positive are indeed positive (low FP)

## F1 -SCore

Precision and recall happens on opposite direction, in order to combine both together and get the combined output we use F1-SCORE

### Confusion Matrix 2*2 using Python

```python
import pandas as pd
import numpy as np
import seaborn as sn

from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
```

## Data Creation for 2 * 2 matrix

```python
actual    = [1, 1, 0, 1, 0, 0, 1, 0, 0, 0] 
predicted = [1, 0, 0, 1, 0, 0, 1, 1, 1, 0] 
```

```python
results = confusion_matrix(actual, predicted) 
results
```

```python
sn.heatmap(results, annot=True)
results
```

```python
pip install matplotlib==3.1.0
```

```python
print ('Accuracy Score :',accuracy_score(actual, predicted) )
# inference : accuracy is 70%
# over all the model performace is 70% good
```

```python
print(classification_report(actual, predicted) )
```

### Multi Classification

```python
from mlxtend.evaluate import confusion_matrix
```

```python
y_target =    [1, 1, 1, 0, 0, 2, 0, 3]
y_predicted = [1, 0, 1, 0, 0, 2, 1, 3]
```

```python
cm = confusion_matrix(y_target=y_target, 
                      y_predicted=y_predicted, 
                      binary=False)
cm
```

```python
sn.heatmap(cm, annot=True)
```

```python
print ('Accuracy Score :',accuracy_score(y_target, y_predicted) )
# inference : accuracy is 75%
# over all the model performace is 75% good
```

```python
print(classification_report(y_target, y_predicted) )
```

#### How Multiclass confussion matrix calculation happens

## recall calculation : for label_a

= TP_A/(TP_A+FN_A)

= TP_A/(Total Gold for A)

= TP_A/TotalGoldLabel_A

= 30/100

= 0.3

## Precision calculation for label_a

 TP_A/(TP_A+FP_A)
 
= TP_A/(Total predicted as A)

= TP_A/TotalPredicted_A

= 30/60

= 0.5

## Recall for label_b

= TP_B/(TP_B+FN_B)

= TP_B/(Total Gold for B)

= TP_B/TotalGoldLabel_B

= 60/100

= 0.6

## Precision for label_b

= TP_B/(TP_B+FP_B)

= TP_B/(Total predicted as B)

= TP_B/TotalPredicted_B

= 60/120

= 0.5
